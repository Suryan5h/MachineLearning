```python
#Lot of specialization, not really useful <10
spec_list = df.Specialization.value_counts()
spec_list

lessThan10 = spec_list[spec_list<=10]
lessThan10

def correctSpec(val):
    if val in lessThan10:
        return 'other'
    else:
        return val
    
df.Specialization = df['Specialization'].apply(correctSpec)
df.Specialization.value_counts()
```

```python
cols_with_nan = [col for col in df.columns if df.isnull().sum()[col]>0]
cols_with_nan
```

```python
for col in cols_with_nan:
    df[col] = df[col].fillna(df[col].mean())
```
